```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1])
```

#### Optimization problem: https://scikit-learn.org/stable/modules/svm.html#mathematical-formulation

```python
C = [0.001, 1, 100]
ratios = [(100, 2), (100, 20), (100, 40), (100, 80)]
plt.figure(figsize = (20, 18))
plt_cont = 1
for j,i in enumerate(ratios):
    X_p = np.random.normal(0, 0.05, size = (i[0], 2))
    X_n = np.random.normal(0.13, 0.02, size = (i[1], 2))
    y_p = np.array([1] * i[0]).reshape(-1, 1)
    y_n = np.array([0] * i[1]).reshape(-1, 1)
    X = np.vstack((X_p, X_n))
    Y = np.vstack((y_p, y_n))
    for k in range(len(C)):
        plt.subplot(len(ratios), len(C), plt_cont)
        clf = SVC(C = C[k])
        clf.fit(X, Y)
        sv = clf.support_vectors_
        plt.scatter(X[:, 0], X[:, 1],c = Y, label = 'C = {}'.format(C[k]))
        
        # Code taken from https://scikit-learn.org/stable/auto_examples/svm/plot_separating_hyperplane.html
        
        # plot the decision function 
        ax = plt.gca()
        xlim = ax.get_xlim()
        ylim = ax.get_ylim()

        # create grid to evaluate model
        xx = np.linspace(xlim[0], xlim[1], 30)
        yy = np.linspace(ylim[0], ylim[1], 30)
        YY, XX = np.meshgrid(yy, xx)
        xy = np.vstack([XX.ravel(), YY.ravel()]).T
        Z = clf.decision_function(xy).reshape(XX.shape)

        # plot decision boundary and margins
        ax.contour(XX, YY, Z, colors = 'k', levels = [-1, 0, 1], alpha = 0.5,
            linestyles = ['--', '-', '--'])
        ax.scatter(sv[:, 0], sv[:, 1], s=100,
            linewidth = 1, facecolors = 'none', edgecolors = 'k', label = 'Support vectors')
        plt.title('Data set {} and regularizer is {}'.format(i, C[k]))  
        plt.legend()      
        plt_cont += 1
plt.suptitle('Effect of Regularization on imbalanced data in case of SVM')      
plt.show()
```

##### Reference: https://machinelearningmastery.com/cost-sensitive-svm-for-imbalanced-classification/

From above graphs we can see that as we increase the value of C, the margin is decreasing. Since `C is inverse of regularization`, for large values of C, overfitting is taking place.

##### Case 1: C = 0.001

It is the case of underfitting, even though the data is visible as linearly separable.

##### Case 2: C = 1

There seems to be have a better margin comparatively.

##### Case 3: C = 100

There is a very thin margin. Even though it could identify support vectors with 100% prediction in all the cases of class imbalance, given a different test data set it might not perform well.

Hence, we conclude that class imbalance does not have much effect, since the margin depends on the support vectors only.

## Applying LR

###### Logistic Regression: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

<pre>these are results we got when we are experimenting with one of the model </pre>
<img src='https://i.imgur.com/vLXTxjY.png'>

```python
C = [0.001, 1, 100]
ratios = [(100, 2), (100, 20), (100, 40), (100, 80)]
plt.figure(figsize = (20, 18))
plt_cont = 1
for j, i in enumerate(ratios):
    X_p = np.random.normal(0, 0.05, size = (i[0], 2))
    X_n = np.random.normal(0.13, 0.02, size = (i[1], 2))
    y_p = np.array([1] * i[0]).reshape(-1, 1)
    y_n = np.array([0] * i[1]).reshape(-1, 1)
    X = np.vstack((X_p, X_n))
    Y = np.vstack((y_p, y_n))
    for k in range(len(C)):
        plt.subplot(len(ratios), len(C), plt_cont)
        clf = LogisticRegression(C = C[k])
        clf.fit(X, Y)
        plt.scatter(X[:, 0], X[:, 1], c = Y ,label = 'C = {}'.format(C[k]))
        draw_line(clf.coef_[0], clf.intercept_, np.min(X[:, 1]), np.max(X[:, 1]))
        plt.legend()
        plt.title('Data set {} and regularizer is {}'.format(i,C[k]))      
        plt_cont += 1
plt.suptitle('Effect of Regularization on imbalanced data in case of Logistic Regression')
plt.show()
```

##### Reference: https://machinelearningmastery.com/cost-sensitive-logistic-regression/

##### Case 1: C = 0.001

- As C is very small, the model is unable to classify the data. 
- We observe that the hyperplane position is away from the data points. 
- Data being balanced or imbalanced doesn't have any difference.

##### Case 2: C = 1

- The hyperplane is far from the data points, we can say that this model cannot classify the imbalanced dataset. 
- In the last case, where the dataset is almost balanced, we see that the model can classify better. It seems to work well only if the data is balanced.

##### Case 3: C = 100

- The model is not able to classify the highly imbalanced dataset as even with a high value of C. 
- Hence, we can conclude that this model does not work well, or it's not recommended to use this when we have a highly imbalanced dataset. 
- As the data becomes balanced, the model does a pretty good job in classification.

### Concept 

- For both Logistic Regression and SVM, `C` is the hyperparameter which is `inverse of regularization`.

- We need to use regularization to avoid overfitting.

- As C increases, $\Lambda$ decreases, there is clear distinction between the classes in the dataset. The separation between the $\Pi$+ and $\Pi$- decreases, average distance of the misclassified points decreases. In other words, it leads to overfitting. It is guaranteed to work well only on the train data.

- As C decreases, $\Lambda$ increases, there is no distinction between the classes in the dataset. The separation between the $\Pi$+ and $\Pi$- increases, average distance of the misclassified points increases. In other words, it leads to underfitting.
