## _Applying Logistic Regression_

#### _1. Correlation between features_

```python
# Check the correlation between the features.
correlation_matrix = data.corr()

print(correlation_matrix);
```

```python
# Plot heat map of correlation matrix using seaborn heatmap.
sns.heatmap(correlation_matrix, annot = True, cmap = "Blues", linewidths = 0.001);

plt.gcf().set_size_inches(10, 10);
```

#### _2. Finding the best model for the given data_

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

alphas = np.logspace(-2, 1, 20).tolist() # choose the alpha in log space
print(alphas)
```

```python
clf_lr = LogisticRegression(random_state = 0)
parameters = {'C': alphas}

# Find the best hyperparameter alpha with hyperparameter tuning using k-fold cross validation - GridSearchCV
clf = GridSearchCV(clf_lr, parameters, cv = 5, scoring = 'accuracy')

# Train Logistic Regression on data(X,Y) 
clf = clf.fit(X, Y)
```

```python
clf.best_params_, clf.best_score_, clf.best_estimator_
```

```python
# Create a new Logistic Regression with the best alpha
best_model = LogisticRegression(random_state = 0, C = 0.01, max_iter = 10);
```

#### _3. Getting the weights with the original data_

```python
# Train the 'best_model' with X, Y
best_model.fit(X, Y);
pred = best_model.predict(X);

# Check the accuracy of the model
best_model_accuracy = accuracy_score(Y, pred);
print("Best model accuracy:", best_model_accuracy);

# Get the weights
W = best_model.coef_
print("Best model weights:", W);
```

#### _4. Modifying original data_

```python
# Add a noise(order of 10^-2) to each element of X and get the new data set X' 
noise = 0.01

X_dash = X + noise
```

```python
fig = plt.figure(figsize = (20, 15))

plt.subplot(3, 1, 1);
plt.plot(X);
plt.title("X Datapoints")

plt.subplot(3, 1, 2);
plt.plot(X_dash);
plt.title("X_dash Datapoints")

plt.subplot(3, 1, 3);
plt.plot(np.subtract(X, X_dash));
plt.title("X-delta Datapoints")

fig.show()
```

```python
# Train the same 'best_model' with data (X', Y)
best_model_edited = LogisticRegression(C = 0.01, random_state = 0, max_iter = 10)
best_model_edited.fit(X_dash, Y)
pred = best_model_edited.predict(X_dash)

# Check the accuracy of the model 
best_model_accuracy_edited = accuracy_score(Y, pred)
print("Best Model accuracy edited:", best_model_accuracy_edited)

print("Best model weights:", best_model.coef_, "\n")

# Get the weights W'
W_dash = best_model_edited.coef_
print("Best model edited-weights:", W_dash)
```

#### _5. Checking deviations in metric \& weights_

```python
# Find the difference between 'best_model_accuracy_edited' and 'best_model_accuracy'
accuracy_diff = best_model_accuracy - best_model_accuracy_edited
print("Difference in model accuracy:", accuracy_diff)
```

```python
# Find the absolute change between each value of W and W' ==> |(W-W')|
weight_changes = np.abs(np.subtract(best_model.coef_, best_model_edited.coef_))
print("Deviations in weights:", weight_changes)

# Print the top 4 features which have higher % change in weights
features = np.argsort(np.subtract(best_model.coef_, best_model_edited.coef_)).flatten().tolist()

print("Top 4 affected features:", [list(data.columns)[i] for i in sorted(features, reverse = True)[:4]])
```

### _Observations_

- After perturbation test, we can see that there is a very slight difference in weights. 

- There is no change in accuracy of the model even after applying perturbation test. 

- Changes in the independent features after perturbation test affects the corresponding collinear feature weights.

## _Applying Linear SVM_

#### _2. Finding the best model for the given data_

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

clf_svm = LinearSVC(random_state = 0)
parameters = {'C': alphas}

# Find the best hyperparameter alpha with hyperparameter tuning using k-fold cross validation - GridSearchCV
clf = GridSearchCV(clf_svm, parameters, cv = 5, scoring = 'accuracy')

# Train Linear SVM on data(X,Y) 
clf = clf.fit(X, Y)
```

```python
clf.best_params_, clf.best_score_, clf.best_estimator_
```

```python
# Create a new SVM with the best alpha
best_model_svm = LinearSVC(random_state = 0, C = 0.01, max_iter = 10);
```

#### _3. Getting the weights with the original data_

```python
# Train the 'best_model' with X, Y
best_model_svm.fit(X, Y)
pred = best_model_svm.predict(X)

# Check the accuracy of the model
best_model_svm_accuracy = accuracy_score(Y, pred);
print("Best model accuracy:", best_model_svm_accuracy);

# Get the weights
W = best_model_svm.coef_
print("Best model weights:", W);
```

#### _4. Modifying original data_

```python
# Add a noise(order of 10^-2) to each element of X and get the new data set X' 
noise = 0.01

X_dash_svm = X + noise
```

```python
fig = plt.figure(figsize = (20, 15))

plt.subplot(3, 1, 1);
plt.plot(X);
plt.title("X Datapoints")

plt.subplot(3, 1, 2);
plt.plot(X_dash_svm);
plt.title("X_dash Datapoints")

plt.subplot(3, 1, 3);
plt.plot(np.subtract(X, X_dash_svm));
plt.title("X-delta Datapoints")

fig.show()
```

```python
# Train the same 'best_model' with data (X', Y)
best_model_svm_edited = LinearSVC(C = 0.01, random_state = 0, max_iter = 10)
best_model_svm_edited.fit(X_dash_svm, Y)
pred = best_model_svm_edited.predict(X_dash_svm)

# Check the accuracy of the model 
best_model_accuracy_svm_edited = accuracy_score(Y, pred)
print("Best Model accuracy edited:", best_model_accuracy_svm_edited)

print("Best model weights:", best_model_svm.coef_, "\n")

# Get the weights W'
W_dash = best_model_svm_edited.coef_
print("Best model edited-weights:", W_dash)
```

#### _5. Checking deviations in metric \& weights_

```python
# Find the difference between 'best_model_accuracy_edited' and 'best_model_accuracy'
accuracy_diff = best_model_svm_accuracy - best_model_accuracy_svm_edited
print("Difference in model accuracy:", accuracy_diff)
```

```python
# Find the absolute change between each value of W and W' ==> |(W-W')|
weight_changes = np.abs(np.subtract(best_model_svm.coef_, best_model_svm_edited.coef_))
print("Deviations in weights:", weight_changes)

# Print the top 4 features which have higher % change in weights
features = np.argsort(np.subtract(best_model_svm.coef_, best_model_svm_edited.coef_)).flatten().tolist()

print("Top 4 affected features:", [list(data.columns)[i] for i in sorted(features, reverse = True)[:4]])
```
